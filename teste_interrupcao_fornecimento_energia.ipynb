```python
import pandas as pd
```

```python
column_names = ['DatGeracaoConjuntoDados', 'IdeConjUndConsumidoras', 'CodMunicipio ', 'NomMunicipio', 'SigUF']
arq_municipios = pd.read_csv('indqual_municipio.csv', encoding='latin-1', delimiter=';', names=column_names, header=None)
arq_municipios.head()
```

<h1> Criar variável com a lista dos Indentificadores do Conjunto de Unidades consumidoras da cidade de São Paulo <h1>

```python
cidade_sp = arq_municipios[arq_municipios['NomMunicipio'] == 'Recife']
id_sp = cidade_sp['IdeConjUndConsumidoras']
lista_id_sp = id_sp.tolist()

```

<h1> Ler dado do ano x de interrupção do fornecimento de energia <h1>


```python
column_names = ["DatGeracaoConjuntoDados","IdeConjuntoUnidadeConsumidora","DscConjuntoUnidadeConsumidora","DscAlimentadorSubestacao","DscSubestacaoDistribuicao","NumOrdemInterrupcao","DscTipoInterrupcao","IdeMotivoInterrupcao","DatInicioInterrupcao","DatFimInterrupcao","DscFatoGeradorInterrupcao","NumNivelTensao","NumUnidadeConsumidora","NumConsumidorConjunto","NumAno","NomAgenteRegulado","SigAgente","NumCPFCNPJ"]
arq_interrupcao = pd.read_csv('interrupcoes-energia-eletrica-2017.csv', encoding='latin-1', names=column_names, delimiter=';',header=None,low_memory=False)
arq_interrupcao.head()
```

<h1> Criar variáveis de interrupção somente para a cidade de São Paulo <h1>

```python
#Transformar as colunas da Tabela arq_interrupcao em lista:
#1)"DatGeracaoConjuntoDados"
lista = arq_interrupcao["DatGeracaoConjuntoDados"]
lista_dat_gcd = lista.tolist()
if lista_dat_gcd[0] == "DatGeracaoConjuntoDados":
  lista_dat_gcd = lista_dat_gcd[1:]
#2)IdeConjuntoUnidadeConsumidora
lista = arq_interrupcao["IdeConjuntoUnidadeConsumidora"]
lista_ide_cuc = lista.tolist()
if lista_ide_cuc[0] == "IdeConjuntoUnidadeConsumidora":
  lista_ide_cuc = lista_ide_cuc[1:]
#3)DscConjuntoUnidadeConsumidora
lista = arq_interrupcao["DscConjuntoUnidadeConsumidora"]
lista_dsc_cuc = lista.tolist()
if lista_dsc_cuc[0] == "DscConjuntoUnidadeConsumidora":
  lista_dsc_cuc = lista_dsc_cuc[1:]
#4)DscAlimentadorSubestacao
lista = arq_interrupcao["DscAlimentadorSubestacao"]
lista_dsc_asub = lista.tolist()
if lista_dsc_asub[0] == "DscAlimentadorSubestacao":
  lista_dsc_asub = lista_dsc_asub[1:]
#5)DscSubestacaoDistribuicao
lista = arq_interrupcao["DscSubestacaoDistribuicao"]
lista_dsc_subd = lista.tolist()
if lista_dsc_subd[0] == "DscSubestacaoDistribuicao":
  lista_dsc_subd = lista_dsc_subd[1:]
#6)NumOrdemInterrupcao
lista = arq_interrupcao["NumOrdemInterrupcao"]
lista_num_oi = lista.tolist()
if lista_num_oi[0] == "NumOrdemInterrupcao":
  lista_num_oi = lista_num_oi[1:]
#7)DscTipoInterrupcao
lista = arq_interrupcao["DscTipoInterrupcao"]
lista_dsc_ti = lista.tolist()
if lista_dsc_ti[0] == "DscTipoInterrupcao":
  lista_dsc_ti = lista_dsc_ti[1:]
#8)IdeMotivoInterrupcao
lista = arq_interrupcao["IdeMotivoInterrupcao"]
lista_ide_mi = lista.tolist()
if lista_ide_mi[0] == "IdeMotivoInterrupcao":
  lista_ide_mi = lista_ide_mi[1:]
#9)DatInicioInterrupcao
lista = arq_interrupcao["DatInicioInterrupcao"]
lista_dat_ii = lista.tolist()
if lista_dat_ii[0] == "DatInicioInterrupcao":
  lista_dat_ii = lista_dat_ii[1:]
#10)DatFimInterrupcao
lista = arq_interrupcao["DatFimInterrupcao"]
lista_dat_fi = lista.tolist()
if lista_dat_fi[0] == "DatFimInterrupcao":
  lista_dat_fi = lista_dat_fi[1:]
#11)DscFatoGeradorInterrupcao
lista = arq_interrupcao["DscFatoGeradorInterrupcao"]
lista_dsc_fgi = lista.tolist()
if lista_dsc_fgi[0] == "DscFatoGeradorInterrupcao":
  lista_dsc_fgi = lista_dsc_fgi[1:]
#12)NumNivelTensao
lista = arq_interrupcao["NumNivelTensao"]
lista_num_nt = lista.tolist()
if lista_num_nt[0] == "NumNivelTensao":
  lista_num_nt = lista_num_nt[1:]
#13)NumUnidadeConsumidora
lista = arq_interrupcao["NumUnidadeConsumidora"]
lista_num_uc = lista.tolist()
if lista_num_uc[0] == "NumUnidadeConsumidora":
  lista_num_uc = lista_num_uc[1:]
#14)NumConsumidorConjunto
lista = arq_interrupcao["NumConsumidorConjunto"]
lista_num_cc = lista.tolist()
if lista_num_cc[0] == "NumConsumidorConjunto":
  lista_num_cc = lista_num_cc[1:]
#15)NumAno
lista = arq_interrupcao["NumAno"]
lista_num_ano = lista.tolist()
if lista_num_ano[0] == "NumAno":
  lista_num_ano = lista_num_ano[1:]
#16)NomAgenteRegulado
lista = arq_interrupcao["NomAgenteRegulado"]
lista_nom_ar = lista.tolist()
if lista_nom_ar[0] == "NomAgenteRegulado":
  lista_nom_ar = lista_nom_ar[1:]
#17)SigAgente
lista = arq_interrupcao["SigAgente"]
lista_sig_ag = lista.tolist()
if lista_sig_ag[0] == "SigAgente":
  lista_sig_ag = lista_sig_ag[1:]
#18)NumCPFCNPJ
lista = arq_interrupcao["NumCPFCNPJ"]
lista_num_cpf = lista.tolist()
if lista_num_cpf[0] == "NumCPFCNPJ":
  lista_num_cpf = lista_num_cpf[1:]

#Criar listas que iram compor a tabela nova:
#1)"DatGeracaoConjuntoDados"
lista_dat_gcd_new = []
#2)IdeConjuntoUnidadeConsumidora
lista_ide_cuc_new = []
#3)DscConjuntoUnidadeConsumidora
lista_dsc_cuc_new = []
#4)DscAlimentadorSubestacao
lista_dsc_asub_new = []
#5)DscSubestacaoDistribuicao
lista_dsc_subd_new = []
#6)NumOrdemInterrupcao
lista_num_oi_new = []
#7)DscTipoInterrupcao
lista_dsc_ti_new = []
#8)IdeMotivoInterrupcao
lista_ide_mi_new = []
#9)DatInicioInterrupcao
lista_dat_ii_new = []
#10)DatFimInterrupcao
lista_dat_fi_new = []
#11)DscFatoGeradorInterrupcao
lista_dsc_fgi_new = []
#12)NumNivelTensao
lista_num_nt_new = []
#13)NumUnidadeConsumidora
lista_num_uc_new = []
#14)NumConsumidorConjunto
lista_num_cc_new = []
#15)NumAno
lista_num_ano_new = []
#16)NomAgenteRegulado
lista_nom_ar_new = []
#17)SigAgente
lista_sig_ag_new = []
#18)NumCPFCNPJ
lista_num_cpf_new = []

#Encontrar selecionar apenas as linhas referentes a cidade de SP:
for i in range(len(lista_ide_cuc)):
  if lista_ide_cuc[i] in lista_id_sp:
    lista_dat_gcd_new.append(lista_dat_gcd[i])
    lista_ide_cuc_new.append(lista_ide_cuc[i])
    lista_dsc_cuc_new.append(lista_dsc_cuc[i])
    lista_dsc_asub_new.append(lista_dsc_asub[i])
    lista_dsc_subd_new.append(lista_dsc_subd[i])
    lista_num_oi_new.append(lista_num_oi[i])
    lista_dsc_ti_new.append(lista_dsc_ti[i])
    lista_ide_mi_new.append(lista_ide_mi[i])
    lista_dat_ii_new.append(lista_dat_ii[i])
    lista_dat_fi_new.append(lista_dat_fi[i])
    lista_dsc_fgi_new.append(lista_dsc_fgi[i])
    lista_num_nt_new.append(lista_num_nt[i])
    lista_num_uc_new.append(lista_num_uc[i])
    lista_num_cc_new.append(lista_num_cc[i])
    lista_num_ano_new.append(lista_num_ano[i])
    lista_nom_ar_new.append(lista_nom_ar[i])
    lista_sig_ag_new.append(lista_sig_ag[i])
    lista_num_cpf_new.append(lista_num_cpf[i])
print(len(lista_dat_gcd_new))
```

<h1> Lista da descrição das subestações e dicionário em que as chaves são as subestações de distribuição <h1>

```python
#Lista das subestações
nome_subestacoes_distribuicao = []
for i in range(len(lista_dsc_subd_new)):
  if i == 0:
    nome_subestacoes_distribuicao.append(lista_dsc_subd_new[i])
  else:
    if lista_dsc_subd_new[i] not in nome_subestacoes_distribuicao:
      nome_subestacoes_distribuicao.append(lista_dsc_subd_new[i])
print(nome_subestacoes_distribuicao)
print(len(nome_subestacoes_distribuicao))
#Dicionario
dic = {}
for i in range(len(nome_subestacoes_distribuicao)):
  chave = nome_subestacoes_distribuicao[i]
  valores = []
  for j in range(len(lista_dsc_subd_new)):
    if lista_dsc_subd_new[j] == chave:
      string = lista_dat_gcd_new[j]+'; '+lista_ide_cuc_new[j]+'; '+lista_dsc_cuc_new[j]+'; '+lista_dsc_asub_new[j]+'; '+lista_dsc_subd_new[j]+'; '+lista_num_oi_new[j]+'; '+lista_dsc_ti_new[j]+'; '+lista_ide_mi_new[j]+'; '+lista_dat_ii_new[j]+'; '+lista_dat_fi_new[j]+'; '+lista_dsc_fgi_new[j]+'; '+lista_num_nt_new[j]+'; '+lista_num_uc_new[j]+'; '+lista_num_cc_new[j]+'; '+lista_num_ano_new[j]+'; '+lista_nom_ar_new[j]+'; '+lista_sig_ag_new[j]+'; '+lista_num_cpf_new[j]
      valores.append(string)
  dic[chave] = valores

```
