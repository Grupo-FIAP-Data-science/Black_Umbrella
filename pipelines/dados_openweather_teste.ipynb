```python
import requests
import pandas as pd
import time

df = pd.read_csv('../dados/distritos_lat_lon.csv')
df.head(3)

# OpenWeatherMap API Key
api_key = "eb27e58eb68d175624e79e4efed521eb"

# Função para obter dados meteorológicos para um par de coordenadas
def get_weather_data(latitude, longitude, api_key):
    # URL para obter dados meteorológicos atuais
    url = f'http://api.openweathermap.org/data/2.5/forecast?lat={latitude}&lon={longitude}&appid={api_key}&units=metric&lang=pt_br'
    
    # Fazendo a requisição para a API
    response = requests.get(url)
    
    # Verificando se a requisição foi bem-sucedida
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Erro ao buscar dados para lat: {latitude}, lon: {longitude}")
        return None

# Loop pelos dados do DataFrame para obter dados meteorológicos
weather_results = []

for index, row in df.iterrows():
    latitude = row['Latitude']
    longitude = row['Longitude']
    district = row['Distrito']
    
    # Obtendo os dados meteorológicos
    weather_data = get_weather_data(latitude, longitude, api_key)
    
    if weather_data:
        # Extraindo alguns dados do JSON retornado
        current_weather = {
            'Distrito': district,
            'Temperatura (°C)': weather_data['list'][0]['main']['temp'],
            'Descrição': weather_data['list'][0]['weather'][0]['description'],
            'Umidade (%)': weather_data['list'][0]['main']['humidity'],
            'Vento (m/s)': weather_data['list'][0]['wind']['speed']
        }
        weather_results.append(current_weather)
    
    # Pausa para evitar sobrecarga na API
    time.sleep(1)

# Convertendo os resultados para um DataFrame
weather_df = pd.DataFrame(weather_results)
weather_df
```
