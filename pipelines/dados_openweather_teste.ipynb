```python
import requests
import pandas as pd

latitude = -23.5505
longitude = -46.6333

# OpenWeatherMap API Key
api_key = "eb27e58eb68d175624e79e4efed521eb"

# URL para obter dados meteorológicos atuais
url = f'http://api.openweathermap.org/data/2.5/forecast?lat={latitude}&lon={longitude}&appid={api_key}'

# Fazer a solicitação GET
response = requests.get(url)

# Verificar o status da resposta
if response.status_code == 200:
    # Converter a resposta JSON para um dicionário Python
    data = response.json()
    
    # Extrair os dados relevantes do JSON
    forecasts = data['list']
    
    # Criar listas para armazenar os dados
    timestamps = []
    temperatures = []
    weather_descriptions = []
    
    for forecast in forecasts:
        timestamps.append(forecast['dt_txt'])
        temperatures.append(forecast['main']['temp'])
        weather_descriptions.append(forecast['weather'][0]['description'])
    
    # Criar um DataFrame com os dados
    df = pd.DataFrame({
        'Timestamp': timestamps,
        'Temperature': temperatures,
        'Weather Description': weather_descriptions
    })
    
    print(df)
else:
    print(f'Erro ao fazer a solicitação: {response.status_code}')
```
