```python
import pandas as pd
from pyproj import Proj, Transformer
```

```python
df = pd.read_excel('../dados/localizacao_servicos_saude.xlsx')
```

```python
df
```

```python
# Defina os sistemas de coordenadas
# Ajuste a zona UTM conforme necessário
utm_proj = Proj(proj='utm', zone=23, south=True, ellps='WGS84')
wgs84_proj = Proj(proj='latlong', datum='WGS84')

transformer = Transformer.from_proj(utm_proj, wgs84_proj)

def converte_para_latlon(x, y):
    # Converter coordenadas UTM para latitude e longitude
    lat, lon = transformer.transform(x, y)
    return lat, lon

# Criar um DataFrame com os resultados
df[['Longitude', 'Latitude']] = df.apply(lambda row: pd.Series(converte_para_latlon(row['COORD_E'], row['COORD_N'])), axis=1)
```

```python
df.to_csv('../dados/localizacao_servicos_saude_latlon.csv', index=False)
```
